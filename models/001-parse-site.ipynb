```python
import requests
from io import BytesIO
from zipfile import ZipFile
import re
```

```python
pages = 1000

for i in range(pages):
    url = 'https://ragnacustoms.com/songs/ddl/' + str(i)
    r = requests.get(url)
    print('Song :', i, r.status_code)
    
    if r.status_code == 200:
        zipfile = ZipFile(BytesIO(r.content))

        for name in zipfile.namelist():
            zipfile.extract(name, 'data/')
```
